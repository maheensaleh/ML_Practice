# Grouping Data 

### grouping data by a specific column is same as applying a specific filter to the df to get a data based on a particular value 
### but for all differnet values of a column like coiuntries we have to defien different different filters to group data by each country name
### whereas group will automatically group data wrt to every value

```python
data["Hobbyist"].value_counts()
```

```python
hobbyist = data.groupby('Hobbyist')
```

```python
hobbyist
type(hobbyist)
```

```python
hobbyist.get_group('Yes')
```

```python
hobbyist["Country"].value_counts().loc["Yes"]

```

## using aggregate functions on groups

hobbyist is a group based on its different values i.e yes and no 
we will perform some aggregate fucntions on some specific columns of these groups

```python
hobbyist['ConvertedComp'].describe()
```

```python
hobbyist.get_group('Yes')['ConvertedComp'].median()
```

```python
# apply a list of specific aggregate functions
hobbyist["ConvertedComp"].agg(['mean','median'])
```

```python
hobbyist["ConvertedComp"].agg(['mean','median']).loc["Yes"]
```
