```python
import numpy as np
import pandas as pd
```

```python
df = {'first':['a','a','d','e','g','m','w','o'],'last':['q','w','e','e','r','t','y','a']}
```

```python
small_df = pd.DataFrame(df)
```

```python
small_df
```

```python
small_df.sort_values('first')
```

```python
small_df.sort_values(by = ['first','last'],ascending=[True,False],inplace=True)
```

```python
small_df
```

```python
small_df.sort_index() # this will again sort them acc to the index
```

```python
small_df
```

## nlargest nsmallest

```python
small_df['middle'] = [1,2,3,4,4,7,5,6]
```

```python
small_df
```

```python
small_df['middle'].nlargest(3)
```

```python
small_df.nlargest(3,'middle')
```
