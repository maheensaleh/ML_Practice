#### make a df from python dic

```python
dic = {"name":['maheen','sara','maryum'],"age":[20,19,18],'resident':['fbarea','gulshan','fbarea'],"phone":['0000','11111','22222']}
dic_df = pd.DataFrame(dic)
display(dic_df)
print("names series")
display(dic_df['name'])
print("resident series")
display(dic_df.resident)
display(dic_df[['name','age']])
```

#### Accessing rows with 'loc' and 'iloc'

```python
display(dic_df.loc[0])
print("---------------")
display(dic_df.loc[1])
```

```python
display(records.loc[0:4])
display(records.loc[0:4,"Country":'EdLevel'])
```

#### and or not operators in filtering

```python
records.columns
```

```python
records.loc[0,['Hobbyist','Student']]
```

```python
filter2 = (records['Hobbyist']=='Yes') & (records['Student']=='No') & (records['CompTotal']>=5000) & (records['Age1stCode']<'18')
filter2
```

```python
records[filter2].shape[0]
```

```python
records.loc[filter2,['Age1stCode','YearsCodePro','YearsCode','CompTotal']]
```

```python
#~ is the not operator
records.loc[~filter2,['Age1stCode','YearsCodePro','YearsCode','CompTotal']]
```

### setting filter with  a list

```python
l = ['United Kingdom','Canada','Thailand']
f = records['Country'].isin(l)
df2 = records.loc[f]
countries = df2['Country']
```

```python
countries.value_counts()
```
