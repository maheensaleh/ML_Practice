## updating rows

```python
# small_df[fil]['year']=2101  <- this error , so use .loc
small_df.loc[fil,'year']=2101
small_df
```

## update mutiple rows at once using : apply - map - applymap - replace

### apply : to apply a certain function

```python
# for a seires
small_df['name'] = small_df['name'].apply(lambda x : x.upper())
small_df
```

```python
# for a df: when apply used on a df, it works on each series
small_df.apply(lambda s : max(s) #-> this gives the length of each series
```

### apply map : works on every value of df

```python
small_df.rename(columns={'name':'amount'},inplace = True)
small_df['amount'] = [33,44,55]
small_df.loc[2,'year']=2099
small_df=small_df.applymap(lambda x : x*2)
small_df
```

### map : works on series to change specific values of it

```python
small_df['amount'].map({66:1,110:3})
# it leaves unchanged as NaN, to not do NaN, use replace
```

### replace

```python
small_df
```

```python
small_df=small_df.replace({66:1,110:3})
```
