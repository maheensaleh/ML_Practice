```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
startup = pd.read_csv("50_Startups.csv")
startup
```

```python
states_to_encode = startup["State"]  # picking teh column which has to be encoded
states_to_encode
```

```python
#assigining labels ot each states name

labelencoder= LabelEncoder()
labeled_states = labelencoder.fit_transform(states_to_encode)
labeled_states
```

```python
startup['State']=labeled_states.reshape(50,1)
startup
#states are now represented as labels
```

```python
#one hot encoding 
onehotencoder = OneHotEncoder(categorical_features = [3])
encoded_data = onehotencoder.fit_transform(startup)
```

```python
# data is now in form of array
data = encoded_data.toarray()
print(type(data))
data
```

```python
#creating data frame from array
edata = pd.DataFrame(data=data[:,:],    # values
              columns=["california","florida","newyork","R&D Spend","Administration","Marketing Spend","profit"])  # 1st row as the column names
```
